```python
"""Quick script for viewing the errors of the mimic 50 rarest multi-label classification notes"""
import pandas as pd
import json

pd.options.display.width = None; pd.options.display.max_rows = 999; pd.set_option('display.max_colwidth', None)

with open("icd_mimic3_random_sort.json", "r") as f:
    codes_to_descriptions = json.load(f)
```

```python
df = pd.read_csv("eval_results.csv")
```

```python
print(len(df[df["correct"] == 0]))
print(len(df[df["correct"] == 1]))
```

```python
errors = df[df["correct"] == 0]
for i, e in errors.iterrows():
    print(f"{'#'*100}")
    try:
        print("##### CORRECT LABEL: ", e.LABELS,"-", codes_to_descriptions[e.LABELS][0], "######")
        print(e["RAW_DISCHARGE_TEXT"])
    except (KeyError, IndexError):
        print("Error")
```
